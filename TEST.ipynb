```python
user_track_matrix = df.pivot_table(index='name',columns='Track', values='count',fill_value=0)
print(user_track_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity=cosine_similarity(user_track_matrix)
user_similarity=pd.DataFrame(user_similarity, index=user_track_matrix.index, columns=user_track_matrix.index)
print(user_similarity)

```

```python
def recommend_songs(user_name, user_similarity, user_track_matrix, top_n=10):
    # Kiểm tra xem người dùng có tồn tại trong dữ liệu không
    if user_name not in user_track_matrix.index:
        return f" '{user_name}' not in data."
    
    # Lấy chỉ số của người dùng
    user_index = user_track_matrix.index.get_loc(user_name)
    
    # Lấy các người dùng tương đồng nhất (loại bỏ chính người dùng đó)
    similar_users = user_similarity.iloc[user_index].sort_values(ascending=False).index[1:top_n+1]
    
    # Lấy các bài hát mà người dùng hiện tại chưa nghe
    unheard_songs = user_track_matrix.loc[user_name][user_track_matrix.loc[user_name] == 0].index
    
    # Tính toán điểm số gợi ý dựa trên sự tương đồng
    recommendations = []
    for song in unheard_songs:
        song_score = 0
        for similar_user in similar_users:
            song_score += user_track_matrix.loc[similar_user, song] * user_similarity.loc[user_name, similar_user]
        recommendations.append((song, song_score))
    
    # Sắp xếp các gợi ý theo điểm số
    recommendations.sort(key=lambda x: x[1], reverse=True)
    
    # Chỉ trả về tên bài hát (không trả về điểm số)
    return [song for song, _ in recommendations[:top_n]]

# Hàm nhập tên người dùng từ bàn phím
def get_recommendations_from_input():
    user_name = input()
    recommendations = recommend_songs(user_name, user_similarity, user_track_matrix)
    
    if isinstance(recommendations, str):
        print(recommendations)
    else:
        print(f"\nTop {len(recommendations)} recommend for {user_name}:")
        for song in recommendations:
            print(f"{song}")

# Chạy hàm để nhập tên người dùng và nhận gợi ý
get_recommendations_from_input()
```
